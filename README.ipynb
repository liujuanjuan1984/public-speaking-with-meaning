我也有话要说 


============





### ——普通人的当众讲话技能 





这是我在 2010 年 6、7 月份的时候写的一个系列。牛人就不用读这个了，这是写给“普通人”看的。





后来博客网站坏掉了，就只能在 archive.org 上找到。想想有点可惜，就重新整理了一下放在这里。





因为当时并未想着成书，所以是写到哪里算哪里，别指望在这里看到一个“完整的体系”——这个系列只是“对某些人有一些帮助”而已。
